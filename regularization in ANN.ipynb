```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.optimizers import Adam
```

```python
x,y = make_moons(100, noise = .25, random_state=51)
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
plt.show()
```

```python
## without regularization

model1 = Sequential()

model1.add(Dense(128,input_dim = 2, activation  ="relu"))
model1.add(Dense(128,activation="relu"))
model1.add(Dense(1,activation = 'sigmoid'))

model1.summary()
```

```python
adam = Adam(learning_rate=.01)
model1.compile(loss = 'binary_crossentropy', optimizer=adam , metrics=['accuracy'])

history1 = model1.fit(x,y, epochs =2000, validation_split= .2, verbose=0)
```

```python
plot_decision_regions(x,y.astype('int'),clf = model1, legend = 2)
plt.xlim(-2,3)
plt.ylim(-1.5,2)
plt.show()
```

```python
## we can see the loss graph also
plt.plot(history1.history['loss'])
plt.plot(history1.history['val_loss'])
```

```python
## Now we are going to use the regularization L2

model2 = Sequential()

model2.add(Dense(128, input_dim = 2, activation ='relu',kernel_regularizer=tensorflow.keras.regularizers.l1(.001)))
model2.add(Dense(128, activation ='relu',kernel_regularizer=tensorflow.keras.regularizers.l1(.001)))
model2.add(Dense(1, activation = 'sigmoid'))
```

```python
model2.summary()
```

```python
adam = Adam(learning_rate=0.01)
model2.compile(loss = 'binary_crossentropy',optimizer=adam, metrics=['accuracy'])

history2 = model2.fit(x,y,epochs = 2000, validation_split=0.2, verbose=0)
```

```python
plot_decision_regions(x,y.astype('int'),clf = model2, legend =2 )
plt.xlim(-2,3)
plt.ylim(-1.5,2)
plt.show()
```

```python
plt.plot(history2.history['loss'])
plt.plot(history2.history['val_loss'])
```

```python
## boxplot

model1_weight_layer1 = model1.get_weights()[0].reshape(256)
model2_weight_layer1 = model2.get_weights()[0].reshape(256)
```

```python
## box plot of first model without regularization
sns.boxplot(model1_weight_layer1)
```

```python
## box plot of model2 with regularization

sns.boxplot(model2_weight_layer1)
```

```python
## we can compare also there pdf plot

sns.distplot(model1_weight_layer1)
sns.distplot(model2_weight_layer1)
```
